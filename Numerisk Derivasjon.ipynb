# Numerisk Derivasjon

## funksjon for Ã¥ derivere



```python
from pylab import *

def f(x):
    return x**2

def fder(f, x, dx=1E-5):
    df = (f(x+dx)-f(x))/dx
    return df

df = fder(f,1)

print(df)

```

# Derivasjon av diskrete data

```python
from pylab import *

t = [5, 10, 15, 30, 50, 60]
s = [115, 190, 390, 600, 1000, 1200]
v = []
v.append(0) 



for i in range(len(s)-1):
    sder = (s[i+1] - s[i])/(t[i+1]-t[i])
    v.append(sder)

plot(t, s)
```

# heis

```python
from pylab import *
data = loadtxt('heis.txt', skiprows=1, delimiter=',')

#print(data)

tid = data[:,0]
hoyde = data[:, 2]

v = [0]
v.append(0)

for i in range(0,46):
    fart = (hoyde[i+1]-hoyde[i])/(tid[i+1]-tid[i])
    v.append(fart)
    
print("Fart: ", fart, "m/s")



plot(tid, v)
title("fartsgraf")
xlabel('tid i sekunder')
ylabel('farten i m/s')
show()
```
