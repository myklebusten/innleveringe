# Datahåndtering og sånt

### Her er et program som leser av en tekst fil og har standraravvik og gjennomsnitt

```python
from pylab import *

data = loadtxt('Testdata.txt',skiprows=1,delimiter=',')

nr = data[:,0]
hoyde = data[:,1]

print(nr, hoyde)

snitt = mean(hoyde)
avvik = std(hoyde)

print("snitt", snitt, "standaravvik", avvik)
```

```python
from pylab import*

x = [1, 2, 3, 4, 5, 6]
y = [3.2, 4.5, 6.4, 7.5, 8.3, 9.1]

a, b = polyfit(x, y, 1)
x_ny = linspace(0, 7, 1000)
y_ny = a*x_ny + b

scatter(x, y, label = "Datapunkter")
plot(x_ny, y_ny, label = "tilpasset kurve")
show()
```
